```python
text = []
point = []
for num in tqdm(range(25)):
    for num in tqdm(range(10)):    
        page_num = """//*[@id="REVIEW"]/div/div[3]/div/div[2]/a[{}]""".format(num+2)
        driver.find_element_by_xpath(page_num).send_keys(Keys.ENTER)
        time.sleep(2)
        
        rw = driver.find_elements_by_css_selector("span._2Xe0HVhCew")
        pt = driver.find_elements_by_css_selector("em._15NU42F3kT")
        time.sleep(2)
    
        for x, i in enumerate(rw,1):
            text.append(i.text)
        
        for x, i in enumerate(pt,1):
            point.append(i.text) 
```
