```python
driver = webdriver.Chrome('./chromedriver')
driver.implicitly_wait(10)
url = "https://smartstore.naver.com/mngcos/products/156490594"
driver.get(url)

text = []
point = []
for num in tqdm(range(9)):
    for num in range(10):      
        rw = driver.find_elements_by_css_selector("span._3QDEeS6NLn")
        pt = driver.find_elements_by_class_name("_15NU42F3kT")
        time.sleep(2)
    
        for x, i in enumerate(rw,1):
            text.append(i.text)
        
        for x, i in enumerate(pt,1):
            point.append(i.text)
            
            
        page_num = """//*[@id="REVIEW"]/div/div[3]/div/div[2]/div/div/a[{}]""".format(num+2)
        driver.find_element_by_xpath(page_num).click()
        print(page_num)
        time.sleep(2)
        
driver.close()
```
